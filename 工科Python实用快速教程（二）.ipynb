pandas在读取Excel的时候已经做出了判断。我们把所有的类型全部规定好再重新加载一次。

````python
iris_df = pd.read_excel('data/iris_raw.xlsx', dtype={'sepallength':'float32', 
                                                     'sepalwidth':'float32', 
                                                     'petallength':'float32',
                                                     'petalwidth':'float32',
                                                    'class':'string'})
iris_df.info()
````

这里边还是有一个特例，class是鸢尾花的种类，现在用的是字符串来表示。但是无论是什么样的人工智能模型都无法直接接收数值以外的类型。

必须把class列转化成数值。先来看看这里有多少个种类。这里需要*groupby\(\)*和*count\(\)*函数，据类，每一类做计数。*count\(\)*函数实际上计算的是每一列中的非空值的个数。这种操作特别类似于数据库。

````python
# 分类计数
iris_df.groupby('class').count()
````

再求一个平均值，看一下品种间的差异。

````python
# 分类求平均值
iris_df.groupby('class').mean()
````

可以看出有三个鸢尾花品种，“Iris-setosa”， “Iris-versicolor”，“Iris-virginica”。
最直观的想法可能是为每个品种给定一个数字，1、2、3，分别代表三种类型。这样做没有数学意义，setosa为什么数值上要比其他两种要小呢？或者说，你在预测的时候遇到了一个结果2.5，它能代表与第2类和第3类相近而与第1类区别等大吗？不行。

在类型的数量非常有限的时候，最常用**独热编码**来替代类型。既然有三个品种，那么加入三个列分别代表是否属于某一个类型，值为“0”代表不属于，为“1”代表属于。我们使用不同的方法加入。

````python
# 加入新列is_setosa
# 使用apply函数
iris_df['is_setosa'] = iris_df['class'].apply(lambda cls: 1 if cls=='Iris-setosa' else 0)

# 查看结果
iris_df.loc[iris_df['class']=='Iris-setosa']
````

````python
iris_df.loc[iris_df['class']!='Iris-setosa']
````

*apply\(\)*就是在每一个经过的元素上施加一个函数，也就是说它是Pandas上的一个高阶函数。小巧玲珑的lambda函数这里用上了。

````python
# 加入is_versicolor列
# 使用map函数传入字典
iris_df['is_versicolor'] = iris_df.loc[:, 'class'].map({'Iris-versicolor':1, 'Iris-virginica':0, 'Iris-setosa':0})

# 查看结果
iris_df.loc[iris_df['class']=='Iris-versicolor']
````

````python
# 加入is_virginica列
# 使用map函数传入字典
iris_df['is_virginica'] = iris_df.loc[:, 'class'].map(lambda cls: 1 if cls == 'Iris-virginica' else 0)

# 查看结果
iris_df.loc[iris_df['class']=='Iris-virginica']
````

现在class列就失去了它的作用了。我们把它删除掉，然后把结果保存起来。

````python
# 删除class列
# 执行多次会报错
iris_df.drop('class', axis=1, inplace=True)
````

*axis=1*表示删除的是列而不是行。*inplace=True*代表删除结果会写入变量*iris_df*。

保存时我们常常不希望保留序列号，也不想要列名，因为准备好的数据应该是给机器读而不再给人去读，不应该有这些。文件会覆盖式的生成到“output”文件夹中。

````python
iris_df.to_csv('output/iris_ready.csv', index=False, header=False)
````

散点图我们就把鸢尾花数据打印出来看看。
鸢尾花有花萼长、宽和花瓣长、宽数据，在前四列，最后三列是is_setosa、is_versicolor、is_virginica三个标志位。

先把花萼长和花瓣长按照二维图分类打印出来。

````python
# 准备数据
def prepare_sizes(class_idx: int, x_idx: int, y_idx: int, iris_array: np.ndarray):
    array_of_class = iris_array[iris_array[:, class_idx]==1]
    X = array_of_class[:, x_idx]
    Y = array_of_class[:, y_idx]
    return X, Y

# setosa
X_sl_setosa, Y_pl_setosa = prepare_sizes(4, 0, 2, iris_nd)

# versicolor
X_sl_versicolor, Y_pl_versicolor = prepare_sizes(5, 0, 2, iris_nd)

# virginica
X_sl_virginica, Y_pl_virginica = prepare_sizes(6, 0, 2, iris_nd)

# 绘图
plt.scatter(X_sl_setosa, Y_pl_setosa, c='r')
plt.scatter(X_sl_versicolor, Y_pl_versicolor, c='g')
plt.scatter(X_sl_virginica, Y_pl_virginica, c='b')

plt.title('Iris classes:')
plt.xlabel('sepal length')
plt.ylabel('patel length')
````

看起来如果我们使用两个长度就可以显示出三个种类不同的规律了。尝试把其他组合也打印出来，而且这回打印在同一个图像的不同子图上。

````python
# 继续准备其他组合
# setosa
X_sl_setosa, Y_pl_setosa = prepare_sizes(4, 0, 2, iris_nd)
X_sl_setosa, Y_sw_setosa = prepare_sizes(4, 0, 1, iris_nd)
X_pl_setosa, Y_pw_setosa = prepare_sizes(4, 2, 3, iris_nd)
X_sw_setosa, Y_sl_setosa = prepare_sizes(4, 1, 2, iris_nd)

# versicolor
X_sl_versicolor, Y_pl_versicolor = prepare_sizes(5, 0, 2, iris_nd)
X_sl_versicolor, Y_sw_versicolor = prepare_sizes(5, 0, 1, iris_nd)
X_pl_versicolor, Y_pw_versicolor = prepare_sizes(5, 2, 3, iris_nd)
X_sw_versicolor, Y_sl_versicolor = prepare_sizes(5, 1, 2, iris_nd)

# virginica
X_sl_virginica, Y_pl_virginica = prepare_sizes(6, 0, 2, iris_nd)
X_sl_virginica, Y_sw_virginica = prepare_sizes(6, 0, 1, iris_nd)
X_pl_virginica, Y_pw_virginica = prepare_sizes(6, 2, 3, iris_nd)
X_sw_virginica, Y_sl_virginica = prepare_sizes(6, 1, 2, iris_nd)

fig = plt.figure()
# subplots的前两个输入参数分别是图像纵向和横向分隔的数量
ax_array = fig.subplots(2, 2, squeeze=False)

# 第一行第一列
ax_array[0, 0].scatter(X_sl_setosa, Y_pl_setosa, c='r')
ax_array[0, 0].scatter(X_sl_versicolor, Y_pl_versicolor, c='g')
ax_array[0, 0].scatter(X_sl_virginica, Y_pl_virginica, c='b')
ax_array[0, 0].set_xlabel('sepal length')
ax_array[0, 0].set_ylabel('petal length')

# 第一行第二列
ax_array[0, 1].scatter(X_sl_setosa, Y_sw_setosa, c='r')
ax_array[0, 1].scatter(X_sl_versicolor, Y_sw_versicolor, c='g')
ax_array[0, 1].scatter(X_sl_virginica, Y_sw_virginica, c='b')
ax_array[0, 1].set_xlabel('sepal length')
ax_array[0, 1].set_ylabel('sepal width')

# 第二行第一列
ax_array[1, 0].scatter(X_pl_setosa, Y_pw_setosa, c='r')
ax_array[1, 0].scatter(X_pl_versicolor, Y_pw_versicolor, c='g')
ax_array[1, 0].scatter(X_pl_virginica, Y_pw_virginica, c='b')
ax_array[1, 0].set_xlabel('petal length')
ax_array[1, 0].set_ylabel('petal width')

# 第二行第二列
ax_array[1, 1].scatter(X_sw_setosa, Y_sl_setosa, c='r')
ax_array[1, 1].scatter(X_sw_versicolor, Y_sl_versicolor, c='g')
ax_array[1, 1].scatter(X_sw_virginica, Y_sl_virginica, c='b')
ax_array[1, 1].set_xlabel('sepal width')
ax_array[1, 1].set_ylabel('sepal length')

plt.suptitle('Iris classes:')
````

## 基本概念

1. 特征向量

在机器学习领域需要先了解“特征”这个概念。每个数据样本就是一组特征，由于特征经常以多个数值的形式存在，就像是空间向量一样，所以特征也会被称为“特征向量”。我们手上的鸢尾花数据每一行的前四列就是特征向量。在复杂的问题里，特征向量的维度可能会有非常大。

2. 模型

所有的特征向量都是为模型准备的。什么是模型？最简单的模型就是二维平面线性回归模型，一个一元一次等式y = ax + b。回归的过程就是求解其中最佳的a和b组合的过程，用它预测就是带入x求y。对于复杂的问题这个等式可能变得非常复杂，可能不再是一个等式而是一个庞大的计算逻辑。但是无论如何它都必须能够接收特征向量作为输入，然后产出一个结果。这个计算逻辑就是模型，它产出的结果叫做预测。

3. 问题类型与模型选择

使用哪种模型需要根据问题类型来决定。机器学习的问题可以按照预测目的分为两种，分类和回归。分类问题的最终目的是根据特征得到离散的类型。我们使用鸢尾花数据来训练模型要做的就是分类。有的时候需要的预测结果是“是”或“否”，也可以算作分类问题。回归问题需要得到的是连续值。分类问题经常可以化为回归问题，因为回归模型输出的值可以被限定在0到1之间，看作是概率。如果按照是否有监督来区别，可以分为有监督和无监督学习。无监督学习使用范围比较有限，一般解决类似分类的问题，但是由于没有标准分类做参考，不算是真正的分类，所以称为聚类。

4. 训练集与验证集

让机器学习的过程被称为训练过程，而将特征输入模型得到结果的过程被称为预测。用于训练模型的数据是训练数据，但是一般为了验证模型的有效性，需要预留一部分数据作为验证数据，不参与训练。更复杂的情况可能要进行多轮训练，一部分数据反复随机划分用于训练和验证，额外在保留一些数据用于训练后的评估。

为了使训练更加高效，数据还需要进行标准化。
